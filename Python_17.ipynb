```python
# Описать иерархию классов проекта 
# Написать тесты к приложению
# Описать исключения к проекту

from pprint import pprint
from math import ceil

# Классы ошибок
class ExceptionCeilingHeight(Exception):
    pass


class ExcludingRoomArea(Exception):
    pass


class ExceptionRoomType(Exception):
    pass


class ExcludingColorTemperature(Exception):
    pass


class Bulb:
    """Родительский класс описания светодиодных ламп""" 
    # описание характеристик лампы и вывод их на экран
    
    def __init__(self, manufacturer, power, light_flow, colorful_temperature): 
        self.manufacturer = manufacturer # производитель
        self.power = power # мощность, Вт
        self.light_flow = light_flow # световой поток, лм
        self.colorful_temperature = colorful_temperature # цветовая температура, К

        self.key = (manufacturer, power) # задание ключа
    
    def __repr__(self): 
        return "Bulb('%s', %s, %s, %s)" % (self.manufacturer, self.power, 
                                           self.light_flow, self.colorful_temperature)
   
    
# Действие 1
# Вычислить количество ламп
# в зависимости от площади помещения и предназначения
# упращенные расчеты без перекрывания, наложения и отражения света (погрешность 10-15%)

class number_lamps(Bulb):
    """Дочерний класс описанcия светодиодных ламп: количество ламп на помещение"""
    def __init__(self, light_flow):
        super().__init__(light_flow)
    
    def amount(self, w, square, height):
        print ('Выберите тип помещения:\n 1- кухня,\n 2 - детская/спальня,\n 3 - ванная/туалет/подсобка,\n 4 - кабинет,\n 5 - лестница,\n 6 - гостинная')
#        w = int(input())
        if w == 1:
            premises = 150 
        elif w == 2:
            premises = 200
        elif w == 3:
            premises = 75
        elif w == 4:
            premises = 300
        elif w == 5:
            premises = 20
        elif w == 6:
            premises = 450
        else:
            raise ExceptionRoomType()
#            raise ValueError ('не верно введен тип помещения')

#        square = int (input('Введите площадь помещения\n'))
        if square <= 0:
            raise ExcludingRoomArea()
            
#        height = float (input('Введите высоту потолков 2.5 - 4.5 м\n'))
        if 2.5 <= height <= 2.7:
            kz1 = 1
        elif 2.7 < height <= 3:
            kz1 = 1.2
        elif 3 < height <= 3.5:
            kz1 = 1.5
        elif 3.5 < height <= 4.5:
            kz1 = 2
        else:
            raise ExceptionCeilingHeight()
#            raise ValueError ('не верно задана высота потолков')
              
        sum_light_flow = premises * square * kz1 * 1.2
        number_lamps =sum_light_flow/self.light_flow
        return print ('количество ламп ', ceil(number_lamps)) 
        
    
# Действие 2
# Выбор цвета лампы:
# видимый свет от 800 К, нижняя границ существующих ламп 2200 К, верхняя - 15000 К
# 2200 - 3400 К - теплый белый
# 3400 - 5000 К - нейтральный белый
# 5000 - 15000 К - холодный белый

class color(Bulb):
    """Дочерний класс описанcия светодиодных ламп: дополняет цвет лампы"""
    def __init__(self, color, colorful_temperature):
        super().__init__(colorful_temperature)
        self.color = color # цвет горящей лампы
    
    def color_light(self):
        if 2200 <= self.colorful_temperature <= 3400:
            self.color = 'теплый белый'
            return print ('цвет света ',self.color, self.colorful_temperature)
        elif 3400 < self.colorful_temperature <= 5000:
            self.color = 'нейтральный белый'
            return print ('цвет света ',self.color, self.colorful_temperature)
        elif 5000 < self.colorful_temperature <= 15000:
            self.color = 'холодный белый'
            return print ('цвет света ',self.color, self.colorful_temperature)
        else:
            raise ExcludingColorTemperature()
#            raise ValueError('не верно задана цветовая температура')
            

if __name__ == '__main__':
    Lamp1 = Bulb('Gauss', 8, 740, 2700)
    Lamp2 = Bulb('Camelion', 13, 1125, 6500)
    Lamp3 = Bulb('Ecola', 9.5, 920, 3200)
    Lamp4 = Bulb('Camelion', 8, 680, 4500)
# с ошибочными данными    
    Lamp5 = Bulb('General', 18, 1680, 1000)
    
    Lamps={
        Lamp1.key: Lamp1,
        Lamp2.key: Lamp2,
        Lamp3.key: Lamp3,
        Lamp4.key: Lamp4,
        Lamp5.key: Lamp5,
    }

# тесты классов    
    print(Bulb.__repr__(Lamp1)) 

# тесты на определение цвета
    Z1 = color.color_light(Lamp1)
    Z2 = color.color_light(Lamp2)
    Z3 = color.color_light(Lamp4)
#    Z4 = color.color_light(Lamp5)

# тесты на количество ламп (характеристики лампы, тип помещения, площадь помещения, высота потолков)
    x1 = number_lamps.amount(Lamp1, 1, 10, 3)
```
