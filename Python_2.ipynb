```python
#Описание с помощью списка

class1=[["Иванов", 2000, 4, 172],
      ["Петров", 2001, 2,164],
      ["Сидоров", 2001, 11, 168]]

# Описание с помощью словаря

Key=["Имя", "год", "месяц", "рост"]

class1_new = dict(zip(Key, class1))


print(class1_new)
print()


class2=[
    {"Имя": "Иванов", "год": 2000,"месяц": 4,"рост": 172},
    {"Имя": "Петров","год": 2001,"месяц": 2,"рост": 164},
    {"Имя": "Сидоров","год": 2001,"месяц": 11,"рост": 168}
    ]


print (class1)
print()
print (class2)





# Список имеет упорядоченную структуру данных, словарь - нет;
# В списках общение по индексу, в словаре - ключ;
# В списках можно выполнять операции над последовательностями, в словарях - нет.
```
