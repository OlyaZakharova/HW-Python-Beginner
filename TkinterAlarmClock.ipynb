```python
# Написать программу будильника
# Будильник срабатывает, когда происходит восход Солнца в выбранном городе


# За основную программу был взят этот будильник
# https://www.askpython.com/python-modules/tkinter/tkinter-alarm-clock
from tkinter import *
import datetime
import time
import winsound
import sys
import ephem
from tkinter import messagebox
 
    
# создание функции timind для отображения часов
def timing():
    # текущий час, минута и секунда
    current_time = time.strftime("%H:%M:%S")
    clock.config(text=current_time)
    # часы будут меняться каждые 210 микросекунд
    clock.after(210,timing)
    
    
def Alarm(set_alarm_timer):
    while True:
        time.sleep(1)
        actual_time = datetime.datetime.now()
        cur_time = actual_time.strftime("%H:%M:%S")
        cur_date = actual_time.strftime("%d/%m/%Y")
        msg="Current Time: "+str(cur_time)
        print (msg)
        if cur_time == set_alarm_timer:
            winsound.PlaySound("Music.wav",winsound.SND_ASYNC)
            break
            

def get_alarm_time():
# на функцию поступают данные введенные пользователем (д, м, г)
# на выходе выдается время рассвета для выбранного города в указанную дату    
    year_2 = int(year.get())
    month_2 = int(month.get())
    day_2 = int(day.get()) 
    print(day_2,month_2,year_2  )
    ac_time = datetime.datetime.now()
# проверка на правильность введения даты
    if int(ac_time.year) > year_2:
        messagebox.showerror("error", "не верно задан год")
        return sys.exit()
    if int(ac_time.month) > month_2:
        messagebox.showerror("error", "не верно задан месяц")
        return sys.exit()
    if int(ac_time.day) > day_2:
        messagebox.showerror("error", "не верно задан день")
        return sys.exit()
           
# Указываем нужный город из библиотеки ephem
# в библиотеке находятся данные для летнго времени
# https://github.com/brandon-rhodes/pyephem/blob/master/ephem/cities.py
    Moscow = ephem.city('Moscow')
    obs = ephem.Observer()
    obs.lat = Moscow.lat #'55.7522222'
    obs.long= Moscow.long #'37.6155556'    
    sun = ephem.Sun()

# Внутри библиотеки ephem  заложены ошибки на введение даты
# дополнительной проверки не требуется
    obs.date = datetime.datetime(year_2, month_2, day_2)

# Время восхода солнца в городе
    rise_time = obs.next_rising(sun)

# Определяем нужные параметры для будильника
    atm = str(ephem.localtime(rise_time).ctime())
    hour = int(atm[11:13])
    minutes = int(atm[14:16])
    seconds = int(atm[17:19])
    
# Проверка полученного времени
    if int(ac_time.hour) > hour:
        messagebox.showerror("error", "рассвет уже наступил")
        return sys.exit()
    elif int(ac_time.hour) == hour:
        if int(ac_time.minute) > minutes:
            messagebox.showerror("error", "рассвет уже наступил")
            return sys.exit()
        elif int(ac_time.minute) == minutes:
            if int(ac_time.second) > seconds:
                messagebox.showerror("error", "рассвет уже наступил")
                return sys.exit()

    alarm_set_time = f"{hour}:{minutes}:{seconds}"
    print (alarm_set_time)
    Alarm(alarm_set_time)
    return atm
    

window = Tk()
window.title("Alarm Clock")
window.geometry("450x300")
window.config(bg="#922B21")
window.resizable(width=False,height=False)

# ОТОБРАЖЕНИЕ Current Time and Date        
setYourAlarm1_1 = Label(window,text = "Current Time and Date: ",fg="white",bg="#922B21",relief = "solid",
                     font=("Helevetica",15,"bold")).place(x=90, y=10) 

addTime1_1 = Label(window,text = "Day     Month     Year",font=60,fg="white",bg="black").place(x = 20, y = 60)  
now = datetime.datetime.now()
year_1 = str(now.year)
month_1 = str(now.month)
day_1 = str(now.day)
setYourAlarm1_2 = Label(window,text = year_1,font=60,fg="yellow",bg="black").place(x = 125, y = 90)
setYourAlarm1_3 = Label(window,text = month_1,font=60,fg="yellow",bg="black").place(x = 75, y = 90)
setYourAlarm1_4 = Label(window,text = day_1,font=60,fg="yellow",bg="black").place(x = 20, y = 90)

addTime1_2 = Label(window,text = "Hour   Min   Sec",font=60,fg="white",bg="black").place(x = 265, y = 50)

# отображение часов
clock=Label(window,font=("times",20,"bold"),bg="yellow")
clock.grid(row=2,column=2,pady=80,padx=270)
timing()

# ОТОБРАЖЕНИЕ Set Time for Alarm
setYourAlarm2 = Label(window,text = "Set Time for Alarm: ",fg="white",bg="#922B21",relief = "solid",
                     font=("Helevetica",15,"bold")).place(x=90, y=130)
addTime2_1 = Label(window,text = "Day    Month    Year",font=60,fg="white",bg="black").place(x = 20, y = 170)
addTime2_2 = Label(window,text = "Hour     Min     Sec",font=60,fg="white",bg="black").place(x = 265, y = 170)
 
day = StringVar()
month = StringVar()
year = StringVar()
 
dayTime= Entry(window, textvariable = day, bg = "#48C9B0", width = 4, font=(20)).place(x=20,y=200)
monthTime= Entry(window, textvariable = month, bg = "#48C9B0", width = 4, font=(20)).place(x=70,y=200)
yearTime = Entry(window, textvariable = year, bg = "#48C9B0", width = 4, font=(20)).place(x=120,y=200)

# кнопка Set Your Alarm
submit = Button(window,text = "Set Your Alarm",fg="Black",bg="#D4AC0D",width = 15,
                command = get_alarm_time,font=(20)).place(x =130,y=240)

# формат времени 24х часовой
time_format=Label(window, text= "Remember to set time in 24 hour format!", fg="white",bg="#922B21",
                  font=("Arial",15)).place(x=20,y=275) 

# запуск программы
window.mainloop()

```
